```python
cols = ['product_id', 'title', 'product_description', 'rating',
        'ratings_count', 'initial_price', 'discount', 'seller_name', 'category']

df = df[cols].copy()
df.rename(columns={
    'title': 'brand',
    'product_description': 'product_name',
    'initial_price': 'price'
}, inplace=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df['seller_name'] = df['seller_name'].fillna('Unknown')
df['rating'] = df['rating'].fillna(df['rating'].median())
df['ratings_count'] = df['ratings_count'].fillna(0)
df['discount'] = df['discount'].fillna(0)

df.dropna(subset=['product_id', 'price', 'category'], inplace=True)
df.reset_index(drop=True, inplace=True)

df.isnull().sum()
```

```python
df.drop_duplicates(subset=['product_id'], keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
np.random.seed(42)
df['quantity'] = np.random.randint(1, 11, size=len(df))
df['total_amount'] = df['price'] * df['quantity']

df[['brand', 'product_name', 'price', 'quantity', 'total_amount']].head(10)
```
